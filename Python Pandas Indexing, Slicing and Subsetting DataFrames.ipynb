# two Indexier function
    - loc()
    - iloc()

```python
df.iloc[0:3,0:4]
```

```python
# select all column for rows of index 0 and 10

df.iloc[[0,10], :]
```

```python
# select specific rows and specific clmuns using iloc()
df.iloc[[3,2,13], [1,4,5]]
```

```python
# select specific rows and specific clmuns using loc() function
# loc function look for colmun names on colmun part
df.loc[[3,2,13], ['Year','Value']]
```

```python
# loc function look for colmun names on colmun part
df.loc[:, 'Year':'Variable_name':2]
```

# Subsetting Data using Criteria
    .We can alse select a subset of our data using criteria.
        -For example, we can select all rows that have a year value of 2002

```python
# Syntax of iloc indexing to finding a specific data
#Remember that python indexing begin at 0,
# So, let index location [2,2] selects the element that is 3 rows down ND 3 columns over in the dataframe
df.iloc[2,2]
df.iloc[2]['Year']
```

```python
df[df.Year==2018]

#Slice between two years
df[(df.Year>2018) & (df.Year<2020)].head(5)
```

```python
df[(df.Year>2018) & (df.Year<2020)].head(5).loc[:, ['Year','Value']]
```

```python
df[(df.Year>2018) & (df.Year<2020)].loc[:, ['Year','Value']].to_csv('datasets/only_2019.csv')
```

# CASE STUDY

```python
unemployment_df = pd.read_csv('datasets/world_unemployment.csv')
```

```python
unemployment_df
```

```python
unemployment_diff_2019_2018= unemployment_df['2019']-unemployment_df['2018']
```

```python
unemployment_df.columns
```

```python
unemployment_sub_df = unemployment_df.loc[::,['Country Name','Country Code','2018', '2019']]
unemployment_sub_df

unemployment_sub_df['2019_2018_diff'] = unemployment_diff_2019_2018
unemployment_sub_df.head(5)
```

```python
# set new index
unemployment_sub_df.set_index('Country Code', append=True, drop=False)
```

```python
# set new index
unemployment_sub_df.set_index('Country Code',drop=False, append=True, verify_integrity=True)
```
